```python
import pandas as pd
from sodapy import Socrata
```

```python
client = Socrata("data.ct.gov", None)
```

```python
results = client.get("5mzw-sjtu", limit=10000000)
```

```python
df = pd.DataFrame.from_records(results)
```

```python
df.head()
```

```python
# returns 10 towns that have the highest sales ratio 
# and 10 towns that have the highest volume of sales 
# for each year available in the dataset.
```

```python
print(type(df["daterecorded"].unique()[0]))
```

```python
df['daterecorded'] = df['daterecorded'].astype('datetime64[ns]')
```

```python
f_dt = df["daterecorded"].unique()[0]
print(f_dt)
```

```python
df = df.groupby(df['daterecorded'].map(lambda x: x.year))
```

```python
df.head()
```

```python
out =  df.groupby(['listyear','town'])['salesratio'].max().reset_index().sort_values(['salesratio'], ascending=[0])
```

```python
out
```

```python
new_df = out.groupby("listyear").head(10).reset_index(drop=True)
```

```python
new_df.loc[new_df['listyear'] == '2001']
```
